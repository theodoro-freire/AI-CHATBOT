```python
# !pip install matplotlib

import matplotlib.pyplot as plt
```

```python
plt.plot(
    [2,9,9,2,2],
    [2,2,9,9,2],
)

plt.title("Quadrado estilizado")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.legend(loc=2)
```

```python
plt.scatter(
    [2,9,9,2,2],
    [2,2,9,9,2],
    label="pontos do cubo",
    c='green',
    edgecolors='pink'
)

plt.title("Quadrado estilizado")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.legend(loc=3, bbox_to_anchor=(1,0.5))
```

```python
plt.figure()

plt.plot(
    [2,9,9,2,2],
    [2,2,9,9,2],
)

plt.scatter(
    [2,9,9,2,2],
    [2,2,9,9,2],
    label="pontos do cubo",
    c='green',
    edgecolors='pink'
)

plt.title("Quadrado estilizado")
plt.xlabel("eixo x")
plt.ylabel("eixo y")
plt.legend(loc=3, bbox_to_anchor=(1,0.5))
plt.savefig("quadrado.pdf")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2 * np.pi, 1024)
data2d = np.log(t)[:, np.newaxis] * np.sqrt(t)[np.newaxis, :]

fig, ax = plt.subplots()
im = ax.imshow(data2d)
ax.set_title('Pan on the colorbar to shift the color mapping\n'
             'Zoom on the colorbar to scale the color mapping')

fig.colorbar(im, ax=ax, label='Interactive colorbar')

plt.show()
```
